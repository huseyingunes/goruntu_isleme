## 🔀 7. Çoklu Değişken Atama

Python'da birden fazla değişkene tek satırda değer atanabilir. İki değişkenin değerleri **tek satırda yer değiştirebilir** — geçici değişken gerekmez!

Bu işleme **tuple unpacking** denir.

````python
a, b, c = 5, "asdf", 7.9
print(a, " : ", b, " : ", c)

# yer değiştirme — geçici değişken olmadan!
a, b = b, a
print(a, " : ", b)  # artık a="asdf", b=5

# aynı değeri birden fazla değişkene atama
x = y = z = "Orange"
print(x)
print(y)
print(z)

y = 15  # sadece y değişti
print(x)
print(y)
print(z)
````

## 🌐 8. Evrensel (Global) Değişkenler

Fonksiyon dışında tanımlanan değişkenler **global** kapsamdadır. Fonksiyon içinden bu değişkeni *değiştirmek* için `global` anahtar kelimesi gerekir; sadece *okumak* için gerekmez.

````python
x = 5

def fonksiyon_():
    y = 4         # sadece bu fonksiyona ait (lokal)
    print(y)

def fonksiyon():
    print(x)      # global x okunabilir
    # print(y)    # HATA! y bu fonksiyonda tanımlı değil

fonksiyon_()
fonksiyon()
````

````python
x = "efsane"

def myfunc():
    global x         # global değişkeni değiştireceğiz
    x = "cok iyi"

myfunc()
print("Python", x)  # "cok iyi"
````

⚠️ Dikkat: `float` → `int` dönüşümünde ondalık kısım **yuvarlanmaz, atılır!**

## 📝 13. Metinler (Strings)

Python'da string'ler `"..."`, `'...'`, ya da üç tırnaklı şekillerle tanımlanabilir.
String'ler **değiştirilemez (immutable)** nesnelerdir.

### String İndeksleme

```
   D  e  n  e  m  e  R
   0  1  2  3  4  5  6   → pozitif indeks
  -7 -6 -5 -4 -3 -2 -1  → negatif indeks
```

````python
a = "DenemeR"

print(a)
print(a[0])      # D
print(a[1])      # e
print(a[2])      # n
print(a[:3])     # Den  (0'dan 3'e kadar, 3 dahil değil)
print(a[3:])     # emeR (3'ten sona)
print(a[2:4])    # ne

print(a[-1])     # R (sondan 1.)
print(a[-4:-2])  # em
print("Metnin uzunluğu: ", len(a))
````

`format()` metodu string içine değişken yerleştirmenin klasik yoludur. Modern Python'da **f-string** daha yaygın ve okunabilirdir.

## 📋 18. Listeler (Lists)

Python listeleri Java/C++'daki array'lerden çok daha güçlüdür:
- **Farklı tipleri** bir arada tutabilir
- **Dinamik boyutludur** — eleman eklenip çıkarılabilir

| Metod | Açıklama |
|-------|----------|
| `append(x)` | Sona ekler |

## 🔃 19. Liste Sıralama

`sort()` metodu listeyi **yerinde** sıralar (listenin kendisini değiştirir). `key` parametresiyle özel sıralama kriteri belirlenebilir. `sorted()` ise yeni bir sıralı liste döndürür, orijinali değiştirmez.

````python
liste = ["Elma", "Armut", "Ayva"]
liste.sort()
print(liste)

liste.reverse()
print(liste)

# Özel sıralama: 50'ye en yakın sayı önce gelsin
def fonksiyon(n):
    return abs(n - 50)

sayi_listesi = [100, 50, 65, 82, 23]
sayi_listesi.sort(key=fonksiyon)
print(sayi_listesi)
````

## 🔗 20. Liste Kopyalama — Referans vs Kopya

⚠️ **Kritik Konu!** Python'da `a = b` bir kopya oluşturmaz — **aynı listeye referans verir!**

Bu Python'a özgü değil — Java, C++, JavaScript gibi diller de referans semantiği kullanır.

````python
sayi_listesi = [3, 5, 10, -8, 9, 15]
a = sayi_listesi    # referans kopyası — aynı nesne!
print(a)

sayi_listesi.sort()  # bu a'yı da değiştirir!
print(a)

print("-" * 40)

# Gerçek kopya için .copy() kullanın:
sayi_listesi = [3, 5, 10, -8, 9, 15]
a = sayi_listesi.copy()   # bağımsız kopya
b = sayi_listesi           # referans
print(a)

sayi_listesi.sort()
print("a (kopya):", a)     # değişmedi
print("b (referans):", b)  # değişti!
````

## 📖 22. Sözlükler (Dictionaries)

Sözlükler **anahtar-değer (key-value)** çiftleri tutar. Python 3.7+'dan itibaren ekleme sırası korunur.

Gerçek hayat benzetmesi: Bir telefon rehberi! İsim (key) → Telefon Numarası (value)

````python
sozluk = {
    "marka": "Ford",
    "model": "Mustang",
    "yil": 1964
}

print(sozluk)
print(sozluk["marka"])
print(sozluk["model"])
print(sozluk["yil"])

# Yeni anahtar ekleme
sozluk["renk"] = "siyah"
print(sozluk)

# Değer güncelleme
sozluk["renk"] = "beyaz"
print(sozluk)

# Anahtarlar ve değerler
print(sozluk.keys())
print(sozluk.values())

# Döngü ile gezme
for i in sozluk.keys():
    print(i, ":", sozluk[i])
````

## 🔢 23. `enumerate()` — İndeksli Döngü

`enumerate()` bir listeyi dönerken hem indeksi hem de değeri aynı anda almanızı sağlar. Manuel sayaç tutmaya gerek kalmaz.

````python
for i, meyve in enumerate(["elma", "armut", "kivi"]):
    print(i + 1, ":", meyve)

# Dikkat çekici bir örnek — ne çıktı verir?
for i, meyve in enumerate(['"a", "b", "c"']):
    print(meyve, i + 1, "-", sep="_")
    break
````

**Görev:** Klavyeden girilen 5 tane sayıyı bir listeye atayarak bunların karelerinden 20 çıktığında ortaya çıkan sonuca göre sıralayan ve listeyi buna göre yazdıran programı yazınız.

**İpucu:** `x = eval(input())`

**Test Verisi:**
```
Girdi:  -5,  3, 10,  1,  7
Kareler: 25,  9, 100,  1, 49
−20:     5,-11,  80,-19, 29
Sıralı: 1, 3, -5, 7, 10
```

````python
# Çözüm 1 — klasik yaklaşım
liste = []
liste.append(eval(input("1. sayıyı giriniz: ")))
liste.append(eval(input("2. sayıyı giriniz: ")))
liste.append(eval(input("3. sayıyı giriniz: ")))
liste.append(eval(input("4. sayıyı giriniz: ")))
liste.append(eval(input("5. sayıyı giriniz: ")))

def fonksiyon(n):
    return n * n - 20

liste.sort(key=fonksiyon)
print(liste)
````

````python
# Çözüm 2 — daha Pythonic (for döngüsü + list comprehension)
sayilar = []
for i in range(5):
    sayi = eval(input(f"{i+1}. sayıyı girin: "))
    sayilar.append(sayi)

# lambda ile de çalışır:
sayilar.sort(key=lambda n: n * n - 20)
print(sayilar)
````
