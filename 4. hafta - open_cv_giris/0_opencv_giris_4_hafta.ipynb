# ⚡ BÖLÜM 2: Neden OpenCV Tercih Edilmeli?

Bir görüntü işleme kütüphanesi seçerken 6 kritik faktör vardır:

## 2.1 Performans — C++ Çekirdeği

OpenCV'nin Python arayüzü kullanıcı dostudur ama altta yatan kod **tamamen C++'ta** yazılmıştır. Bu, Python'ın yavaşlığını ortadan kaldırır:

```
Python kodu
    ↓
OpenCV Python binding (Cython)
    ↓
OpenCV C++ core
    ↓
CPU'ya özel optimize edilmiş kod (SSE/AVX/NEON)
    ↓
Donanım (CPU/GPU)
```

**SIMD Optimizasyonu:** OpenCV, işlemcinin paralel işlem komutlarını (SSE4, AVX2, NEON) otomatik olarak kullanır. Bir döngü yazmak yerine 8-16 piksel aynı anda işlenir.

<div style="background: #001e2d; border-left: 5px solid #00b4d8; padding: 18px 24px; border-radius: 0 10px 10px 0; margin: 14px 0;"><strong style="color: #00b4d8; font-size: 1.08em;">🔬 Teknik Derinlik</strong><br><br><span style="color: #cdd6f4; line-height: 1.9;"><b>SIMD (Single Instruction, Multiple Data) Nedir?</b><br>Modern işlemciler aynı anda birden fazla veri üzerinde aynı işlemi yapabilir. Örneğin AVX2 komutları 256 bit = 32 byte veriye aynı anda işlem uygular. 256x256 piksellik gri tonlamalı bir görüntü üzerinde parlaklık artırma için:<br>• <b>Saf Python döngüsü:</b> 65,536 işlem, tek tek<br>• <b>NumPy:</b> vectorize edilmiş, yaklaşık 8-16x hızlı<br>• <b>OpenCV C++ + SIMD:</b> aynı anda 32 piksel, yaklaşık 50-100x hızlı<br>Gerçek zamanlı video işlemede bu fark hayati önem taşır!</span></div>

## 4.2 OpenCV'de Görüntü Nedir?

OpenCV'de bir görüntü aslında bir **NumPy dizisidir!**

```
Renkli görüntü (BGR):     shape = (yükseklik, genişlik, 3)
Gri tonlamalı görüntü:    shape = (yükseklik, genişlik)
Saydam görüntü (BGRA):    shape = (yükseklik, genişlik, 4)
```

**Piksel değerleri:** uint8 → 0-255 arası tam sayı

⚠️ **Kritik Fark:** OpenCV renk sırasını **BGR** kullanır, RGB değil!
Bu, Intel'in erken görüntü işleme standartlarından miras kaldı.

<div style="background: #1a0d2e; border-left: 5px solid #c77dff; padding: 18px 24px; border-radius: 0 10px 10px 0; margin: 14px 0;"><strong style="color: #c77dff; font-size: 1.08em;">🧠 BGR mi RGB mi? — Tarihin Mirası</strong><br><br><span style="color: #cdd6f4; line-height: 1.9;">OpenCV'nin BGR renk sırası, 1990'lardaki donanım standartlarından gelir. O dönemde Windows bitmap formatı ve bazı kamera donanımları Blue-Green-Red sırasında veri depoluyordu. OpenCV bu standardı benimsedi. Bugün matplotlib, Pillow ve çoğu kütüphane RGB kullanır, bu nedenle OpenCV görüntülerini diğer kütüphanelerle kullanırken <code>cv2.cvtColor(img, cv2.COLOR_BGR2RGB)</code> dönüşümü gerekebilir.<br><br>İlginç not: insan gözü de mavi ışığa en az hassastır — BGR'nin arka planında bir ergonomi mantığı var!</span></div>

## 5.2 Görüntü Okuma — `cv.imread()`

`imread()` fonksiyonu bir görüntü dosyasını okur ve NumPy dizisi olarak döndürür.

```python
cv.imread(dosya_yolu, bayrak)
```

| Bayrak | Değer | Açıklama |
|--------|-------|----------|
| `cv.IMREAD_COLOR` | 1 (varsayılan) | BGR renkli, alpha kanalı yok |
| `cv.IMREAD_GRAYSCALE` | 0 | Gri tonlamalı |
| `cv.IMREAD_UNCHANGED` | -1 | Dosyayı olduğu gibi oku (alpha dahil) |
| `cv.IMREAD_ANYDEPTH` | 2 | 16/32-bit görüntüler için |
| `cv.IMREAD_ANYCOLOR` | 4 | Hangi kanal varsa oku |

````python
import cv2 as cv
import numpy as np

# BGR renkli okuma (varsayılan)
resim = cv.imread("resim/manzara.jpg")

print("Tip:", type(resim))
print("Shape:", resim.shape)
# shape = (yükseklik, genişlik, kanal_sayısı)
# örn: (720, 1280, 3) → 720px yüksek, 1280px geniş, 3 kanal (BGR)

print("veri tipi:", resim.dtype)    # uint8 (0-255)
print("boyut (byte):", resim.nbytes)  # toplam bellek

# İlk pikselin BGR değeri
print("Sol üst piksel (BGR):", resim[0, 0])

cv.imshow("Manzara", resim)
cv.waitKey(0)
cv.destroyAllWindows()
````

<div style="background: #001e2d; border-left: 5px solid #00b4d8; padding: 18px 24px; border-radius: 0 10px 10px 0; margin: 14px 0;"><strong style="color: #00b4d8; font-size: 1.08em;">🔬 Teknik Derinlik</strong><br><br><span style="color: #cdd6f4; line-height: 1.9;"><b>waitKey() Neden Gerekli?</b><br><code>cv.waitKey(n)</code> n milisaniye bekler ve basılan tuşun ASCII kodunu döndürür. <code>n=0</code> sonsuz bekleme anlamına gelir (tuşa basılana kadar).<br><br>OpenCV'nin GUI sistemi (highgui) bir mesaj döngüsüne ihtiyaç duyar. <code>waitKey()</code> bu döngüyü tetikler — çağrılmazsa pencere donabilir veya hiç görünmeyebilir. Bu, işletim sisteminin pencere olaylarını (paint, resize, mouse) işlemesi için zorunludur.<br><br><b>Önemli:</b> Video oynatmada <code>waitKey(33)</code> kullanılır → 33ms ≈ 1000ms/30fps → yaklaşık 30 FPS hız.</span></div>

## 5.3 Gri Tonlamalı Okuma

Gri tonlamalı görüntüler:
- **3x daha az bellek** kullanır (1 kanal vs 3 kanal)
- Birçok algoritma zaten gri tonlamalı çalışır (Canny, threshold, Hough)
- Renk bilgisi önemli değilse her zaman gri kullanın

**Gri dönüşüm formülü:**
```
Gray = 0.299×R + 0.587×G + 0.114×B
```
İnsan gözünün yeşile daha duyarlı, maviye daha az duyarlı olmasından kaynaklanır!

````python
import cv2 as cv

resim = cv.imread("resim/manzara.jpg", cv.IMREAD_GRAYSCALE)

cv.imshow("Baslik", resim)
cv.waitKey(0)

print(resim)          # 2D dizi — sadece parlaklık değerleri
print(resim.shape)    # (yükseklik, genişlik) — kanal yok!
print(resim.ndim)     # 2
print(resim.dtype)    # uint8
````

Gri formülü bunu yansıtır. TV yayıncılığında kullanılan ITU-R BT.601 standardından gelir. Bu yüzden yeşil bir yaprak gri görüntüde parlak, mavi gökyüzü karanlık görünür.</span></div>

## 5.4 Pencere Yönetimi — `imshow()`, `waitKey()`, `destroyWindow()`

OpenCV'nin pencere sistemi basit ama güçlüdür:

| Fonksiyon | Açıklama |
|-----------|----------|
| `cv.imshow(başlık, görüntü)` | Görüntüyü pencerede göster |
| `cv.waitKey(ms)` | ms milisaniye bekle, tuş kodunu döndür |
| `cv.destroyWindow(başlık)` | Belirli pencereyi kapat |
| `cv.destroyAllWindows()` | Tüm pencereleri kapat |
| `cv.namedWindow(başlık, bayrak)` | Önceden pencere oluştur |

## 5.6 Şeffaf Görüntü — Alpha Kanalı

PNG formatı **4. kanal** olarak **alpha** (şeffaflık) bilgisi taşıyabilir:

```
BGR  → 3 kanal: Mavi, Yeşil, Kırmızı
BGRA → 4 kanal: Mavi, Yeşil, Kırmızı, Alpha (0=tamamen şeffaf, 255=tamamen opak)
```

`cv.IMREAD_UNCHANGED` ile okumak bu alpha kanalını korur.

````python
import cv2 as cv
import numpy as np

src = cv.imread("resim/bilgisayar.png", cv.IMREAD_UNCHANGED)

"""
cv.imshow() kullanırken gördüğünüz durum normaldir.
Çünkü OpenCV pencere sistemi alpha (şeffaflık)
kanalını desteklemez. Yani PNG'deki RGBA görüntü
okunur ama gösterilirken alpha kanalı dikkate alınmaz.
"""
print("Seffaf Resim Boyutu :", src.shape)
cv.imshow("Baslik", src)
cv.waitKey(0)
print(src)
````

````python
import cv2 as cv
import numpy as np

# Alpha kanalını kullanarak şeffaf görüntüyü arka planla birleştirme
src = cv.imread("resim/bilgisayar.png", cv.IMREAD_UNCHANGED)

if src.shape[2] == 4:  # alpha kanalı var mı kontrol et
    bgr = src[:, :, :3]          # ilk 3 kanal: renk
    alpha = src[:, :, 3] / 255.0  # alpha: 0.0 - 1.0

    h, w = bgr.shape[:2]
    background = np.ones((h, w, 3), dtype=np.uint8) * 255  # beyaz arka plan

    # Alpha blending formülü:
    # sonuç = ön_plan × alpha + arka_plan × (1 - alpha)
    for c in range(3):
        background[:,:,c] = bgr[:,:,c] * alpha + background[:,:,c] * (1 - alpha)

    cv.imshow("image", background)
    cv.waitKey(0)
else:
    print("Bu görüntüde alpha kanalı yok")
````

<div style="background: #1a0d2e; border-left: 5px solid #c77dff; padding: 18px 24px; border-radius: 0 10px 10px 0; margin: 14px 0;"><strong style="color: #c77dff; font-size: 1.08em;">🧠 Alpha Blending — Sinema ve Oyunların Temeli</strong><br><br><span style="color: #cdd6f4; line-height: 1.9;">Alpha blending formülü <b>Porter-Duff (1984)</b> tarafından geliştirildi ve o günden bu yana grafik dünyasının temel taşı oldu.<br><br><code>sonuç = kaynak × alpha + hedef × (1 - alpha)</code><br><br>Bu basit formül:<br>• Bilgisayar oyunlarında sprite'ları arka plana birleştirmek için<br>• Film görsel efektlerinde CGI karakterleri gerçek çekime eklemek için<br>• Web'de PNG logolarını herhangi arka plana yerleştirmek için<br>• Tıbbi görüntülemede farklı tarama katmanlarını üst üste koymak için kullanılır.<br><br>Saniyede milyarlarca kez çalışan bu küçük formül, modern görsel medyanın gizli kahramanıdır!</span></div>

# 🔬 BÖLÜM 6: Piksel İşlemleri — Derinlemesine

## 6.1 Koordinat Sistemi

OpenCV'de koordinatlar matematikten farklıdır:

```
Matematik:          OpenCV (NumPy):
  y                   [satır, sütun] = [y, x]
  ↑                   Görüntü[0,0] = sol üst
  |                   
  └──→ x              (0,0) ──────────────→ x (sütun)
                        |  [0,0] [0,1] [0,2]
                        |  [1,0] [1,1] [1,2]
                        ↓  [2,0] [2,1] [2,2]
                      y (satır)
```

⚠️ **Kritik:** `resim[y, x]` ama `cv.rectangle(resim, (x1,y1), (x2,y2))` — fonksiyonlar (x,y) alır!

````python
import cv2 as cv
import numpy as np

# Sıfırdan görüntü oluşturma ve piksel işlemleri
resim = np.zeros((300, 400, 3), dtype=np.uint8)

# Tek piksel ayarla
resim[100, 200] = [0, 255, 0]   # (satır=100, sütun=200) → yeşil piksel

# Dikdörtgen bölge doldur
resim[50:150, 50:200] = [255, 0, 0]   # mavi dikdörtgen
resim[50:150, 200:350] = [0, 0, 255]  # kırmızı dikdörtgen

# Tüm satır
resim[200, :] = [255, 255, 0]  # sarı yatay çizgi (BGR: [0,255,255] = sarı değil!)
# NOT: BGR'de sarı = [0, 255, 255], çünkü sarı = yeşil + kırmızı

cv.imshow("piksel islem", resim)
cv.waitKey(0)
````
